# Puff Reader Demo
This notebook demonstrates a function to read the S-parameters from .PUF files generated by the PUFF Microwave CAD software. As an example, the magnitude and phase from a simulation of a design of a branchline coupler are extracted from the file and plotted.

<img src="./Files/BranchlineCoupler_PUFF.png">

```python
%matplotlib inline
import matplotlib.pyplot as plt
import PUFF_Reader

Plots = PUFF_Reader.PUFF_SParameters("./Files/BranchlineCoupler.PUF", dB = True, Radians = False)
plt.figure(figsize=(15, 6))
plt.plot(Plots['Frequencies'], Plots['S11_M'])
plt.plot(Plots['Frequencies'], Plots['S21_M'])
plt.plot(Plots['Frequencies'], Plots['S31_M'])
plt.plot(Plots['Frequencies'], Plots['S41_M'])

plt.title("Branchline Coupler Simulation (Magnitude)", fontsize = 20)
plt.xlabel("Frequency (GHz)")
plt.ylabel("Magnitude (dB)")
plt.grid()
plt.legend(['S11 Magnitude', 'S21 Magnitude', 'S31 Magnitude', 'S41 Magnitude'])

plt.axis([0, max(Plots['Frequencies']), -25, 0])

plt.figure(figsize=(15, 6))
plt.plot(Plots['Frequencies'], Plots['S11_P'])
plt.plot(Plots['Frequencies'], Plots['S21_P'])
plt.plot(Plots['Frequencies'], Plots['S31_P'])
plt.plot(Plots['Frequencies'], Plots['S41_P'])

plt.title("Branchline Coupler Simulation (Phase)", fontsize = 20)
plt.xlabel("Frequency (GHz)")
plt.ylabel("Phase (Degrees)")
plt.grid()
plt.legend(['S11 Phase', 'S21 Phase', 'S31 Phase', 'S41 Phase'])

plt.axis([0, max(Plots['Frequencies']), -190, 190])
```
