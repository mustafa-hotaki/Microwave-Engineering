# S2P Reader Demo
This notebook demonstrates a function to read the S-parameters from S2P Touchstone files. This function does not follow an official reference for Touchstone files. It assumes the beginning and end of the S-parameter information in the file. It also assumes the S-parameters are given in terms of real and imaginary parts. As an example, measurments from a microstrip branchline coupler are extracted from the file and plotted.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import S2P_Reader

Plots = S2P_Reader.S2P_SParameters("./Files/BranchlineCoupler.S2P", dB = True, Radians = False)

plt.figure(figsize=(15, 6))
plt.plot(Plots['Frequencies'], Plots['S11_M'])
plt.plot(Plots['Frequencies'], Plots['S21_M'])
plt.plot(Plots['Frequencies'], Plots['S12_M'])
plt.plot(Plots['Frequencies'], Plots['S22_M'])

plt.title("Branchline Coupler Measurement (Magnitude)", fontsize = 20)
plt.xlabel("Frequency (GHz)")
plt.ylabel("Magnitude (dB)")
plt.grid()
plt.legend(['S11 Magnitude', 'S21 Magnitude', 'S31 Magnitude', 'S41 Magnitude'])

plt.axis([0, max(Plots['Frequencies']), -25, 0])

plt.figure(figsize=(15, 6))
plt.plot(Plots['Frequencies'], Plots['S11_P'])
plt.plot(Plots['Frequencies'], Plots['S21_P'])
plt.plot(Plots['Frequencies'], Plots['S12_P'])
plt.plot(Plots['Frequencies'], Plots['S22_P'])

plt.title("Branchline Coupler Measurement (Phase)", fontsize = 20)
plt.xlabel("Frequency (GHz)")
plt.ylabel("Phase (Degrees)")
plt.grid()
plt.legend(['S11 Phase', 'S21 Phase', 'S31 Phase', 'S41 Phase'])

plt.axis([0, max(Plots['Frequencies']), -190, 190])
```
